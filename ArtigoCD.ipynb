```python
#Imports Necessários
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics

%matplotlib inline
```

```python
#Carregando dados dos APOSENTADOS
path = r'Aposentados'
all_files = glob.glob(path + "\*.csv")
cols = ['Nome','CPF','Matricula','Nome_orgão','Sigla_orgao','Codigo_orgao','Cargo', 'Classe', 'Padrao', 
        'Referencia', 'Nivel' ,'Tipo_Aposentadoria','Fundamentacao_inatividade','Nome_Diploma_Legal',
        'Data_Diploma_Legal','Ocorrencia_Ingresso','Data_ocorrencia ','Rendimento_liq']
dfs = []

for i in range(len(all_files)):
    dfs.append(pd.read_csv(all_files[i], header=None, names=cols,encoding='Latin1' ,sep=';',error_bad_lines=False ))

df_aposentados = pd.concat(dfs) # --> DataFrame de aposentados
df_aposentados
```

```python
#Carregando dados dos ABONOS
path = r'Abonos'
all_files = glob.glob(path + "/*.csv")
cols = ['Nome','CPF','Descrição','Nível','Denominação do órgão de atuação','UF da UPAG de vinculação','Denominação unidade organizacional','UF',
'Cidade', 'Situação', 'anos_serviço_publico','meses_servico_publico','Inicio_Abono','valor']
dfs = []

for i in range(len(all_files)):
    dfs.append(pd.read_csv(all_files[i], header=None, names=cols,encoding='Latin1' ,sep=';',error_bad_lines=False ))

df_abonos = pd.concat(dfs) # --> DataFrame de abonos
df_abonos = df_abonos[df_abonos['Nome']!='Nome']

df_abonos
```

<h1> PROCESSO DE ETL </h1>

```python
#ABONOS

#ABONOS
#Separando tempo
ano_mes = list(df_abonos['Inicio_Abono'])
print(len(ano_mes))
matriz = []
for elemento in ano_mes:
    matriz.append(str(elemento))

meses = list()
anos = list()
for i in range(0, len(ano_mes)):
    meses.append(matriz[i][4]+matriz[i][5])
    anos.append(matriz[i][0]+matriz[i][1]+matriz[i][2]+matriz[i][3])
    
df_abonos['MÊS'] = meses
df_abonos['ANO'] = anos
```

```python
#Transformando valores de String para Float
df_abonos['Denominação do órgão de atuação'] = df_abonos['Denominação do órgão de atuação'].str.strip()
df_abonos['valor'] = df_abonos['valor'].str.replace(' ', '')
df_abonos['valor'] = df_abonos['valor'].str.replace(',', '.')
aux = list(df_abonos['valor'])
for i in range(0, len(aux)):
    aux[i] = float(aux[i])
        
df_abonos['valor'] = aux
```

```python
#Transformando Meses de String para Float
df_abonos['MÊS'] = df_abonos['MÊS'].str.replace(' ', '')
aux = list(df_abonos['MÊS'])
for i in range(0, len(aux)):
    aux[i] = int(aux[i])
        
df_abonos['MÊS'] = aux
```

```python
#Transformando Anos de String para Float
df_abonos['ANO'] = df_abonos['ANO'].str.replace(' ', '')
aux = list(df_abonos['ANO'])
for i in range(0, len(aux)):
    aux[i] = int(aux[i])
        
df_abonos['ANO'] = aux
```

```python
#Transformando anos_serviço_publico de String para Float
df_abonos['anos_serviço_publico'] = df_abonos['anos_serviço_publico']
aux = list(df_abonos['anos_serviço_publico'])
for i in range(0, len(aux)):
    aux[i] = int(aux[i])
        
df_abonos['anos_serviço_publico'] = aux
#df_abonos
```

```python
#Transformando meses_servico_publico de String para Float
df_abonos['meses_servico_publico'] = df_abonos['meses_servico_publico']
aux = list(df_abonos['meses_servico_publico'])
for i in range(0, len(aux)):
    aux[i] = int(aux[i])
        
df_abonos['meses_servico_publico'] = aux
df_abonos
```

```python
df_aposentados['Rendimento_liq'] = df_aposentados['Rendimento_liq'].str.replace('.', '')
df_aposentados['Rendimento_liq'] = df_aposentados['Rendimento_liq'].str.replace(',', '.').astype(float)

df_aposentados.info()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
import os
import pandas
import sklearn
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
df_abonos.sort_values('ANO', ascending=False, inplace=True)
```

```python
df_abonos.drop_duplicates(subset='Nome', keep='first',inplace=True)
```

```python
#calcular o tempo de ingresso no serviço público em anos
ano_inicial = df_aposentados['Data_ocorrencia '].tolist()
ano_aposent = df_aposentados['Data_Diploma_Legal'].tolist()
```

```python
#Tratamento para criação da variável Target Y

def converteAno(x):
    x = str(x)
    x = list(x)
    x = x[-4::]
    x = ''.join(x)
    x = int(x)
    return x

ai = []
af = []

for i in range(len(ano_inicial)):
    ai.append(converteAno(ano_inicial[i]))
    af.append(converteAno(ano_aposent[i]))

ai = np.array(ai)
af = np.array(af)

anos = af - ai

#TARGET
df_aposentados['anos_servico'] = list(anos)
```

```python
#imports
from category_encoders.one_hot import OneHotEncoder, OrdinalEncoder
```

```python
#preparando os dataframes para aplicação do ML
df_1 = df_aposentados.drop(df_aposentados.columns.difference(['Cargo','Nome_orgão','anos_servico']),1,inplace=False)
df_2 = df_abonos.drop(df_abonos.columns.difference(['Descrição','Denominação do órgão de atuação']),1,inplace=False)

df_1.columns = ['orgao','cargo','anos_serv']
df_2.columns = ['cargo','orgao']

ord_enc  = OrdinalEncoder(cols=['cargo','orgao'])
df_train = ord_enc.fit_transform(df_1)
df_test  = ord_enc.fit_transform(df_2)
```

```python
df_test
```

```python
#implementação de rede neural regressora

from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold


X = df_train.drop(['anos_serv'], axis=1)
y = df_train['anos_serv']

kf = StratifiedKFold(n_splits=5) #, shuffle = True, random_state = 2 )
kf.get_n_splits(X)

regr = svm.SVC()

#Execução das rotinas de treino e teste via validação cruzada
for train_index, test_index in kf.split(X,y):
    #iteração dos elementos de treino e teste
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]   
    #ajuste do modelo
    regr.fit(X_train,y_train)
    y_pred = regr.predict(X_test)

pred = regr.predict(X_test)
```

```python
test = df_test.to_numpy()
pred = regr.predict(test)
```

```python
df_abonos['previsao_ano'] = pred
sns.histplot(data=df_abonos, x='previsao_ano')
```

```python
trabalhado = np.array(df_abonos['anos_serviço_publico'].tolist())
tempo_restante = pred - trabalhado
df_abonos['tempo_restante'] = tempo_restante
```

```python
# Histograma de Tempo Estimado até aposentadoria
graph = sns.displot(data=df_abonos, x='tempo_restante', kde=True, bins=20, height=5, aspect=2.5)
plt.ylim(0,22000)
plt.xlim(-25, 40)

plt.xlabel('Tempo estimado até Aposentadoria')
plt.ylabel('Quantidade')

plt.show()


graph.savefig('Histograma de Tempo estimado até Aposentadoria (Geral).png', format='png',dpi=100)
```
